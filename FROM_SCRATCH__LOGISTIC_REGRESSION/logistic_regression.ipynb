```python
def modele(W,x) :
    return W[0]*x + W[1]


def sigmoide(W,x) :
   
    v=1+np.exp(-modele(W,x))
    return 1/v


def logit(p) :
    return np.log(p/(1-p))



def cout(X,y,W):
    s=0
    for i in range(len(y)) :
        #print("log = "+str(sigmoide(W,X[i])))
        #print("yi = "+ str(y[i]))
        s=s+y[i]*np.log(sigmoide(W,X[i])) + (1-y[i])*np.log(1-sigmoide(W,X[i]))
        
        #print("s = "+ str(s))
    return -s/(len(y))


def gradient(X,y,W) :
    sig = []
    for i in range(len(X)):
        sig.append(sigmoide(W,X[i]))

    sig_y=sig-y

    X=np.c_[ X, np.ones(len(X)) ] 
    print(sig_y)
    #X=np.transpose(X)
    sig_y=np.transpose(sig_y)
    result= sig_y @ X 
    return (1/len(y))*result
    
    

def MiseJour(grad,W,alpha):
    W=W-alpha*grad
    return W


def Batch_Gradient_Descent(X,y,W,Nmax,alpha) :
    ms=[]
    a=[]
    b=[]
    for i in range(Nmax) :
        a.append(W[0])
        b.append(W[1])
        ms.append(cout(X,y,W))
        #print(cout(X,y,W))
        
        W=MiseJour(gradient(X,y,W),W,alpha)
       
    return a,b,ms
```
