```python
def fcout(X,y,W) :
    W=W.reshape(len(W),1)
    XW=X @ W
    WX_Y=XW - y
    return WX_Y

def MSE(X,y,W) :
    s=0
    WX_Y=fcout(X,y,W)
    
    for i in range(len(y)) :
        s=s+WX_Y[i]
    return s/(2*len(X))


W = np.array([2, 3, 4, 2])

MSE(X_train,y_train,W)
```
